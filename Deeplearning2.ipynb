```python
import numpy as np
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def softmax(x):
    e = np.exp(x)
    return e / np.sum(e)
```

```python
d = 3
```

```python
Wx = np.random.randn(d, d)
Wh = np.random.randn(d, d)
Wy = np.random.randn(d, d)
```

```python
h = np.zeros((d,1))
```

```python
x_d = np.random.randn(d,1)
x_o = np.random.randn(d,1)
x_g = np.random.randn(d,1)
```

```python
h1 = tanh(Wx @ x_d + Wh @ h)
h2 = tanh(Wx @ x_o + Wh @ h1)
h3 = tanh(Wx @ x_g + Wh @ h2)
```

```python
y = softmax(Wy @ h3)
```
